```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
plt.plot(t, s)
plt.xlabel('time in seconds')
plt.ylabel('voltage in mV')
plt.title('simple plot as example')
plt.grid(True)
plt.show()
```
